<h1 style="font-family:Courier;font-weight:bold;font-size:30px;">Chapter11:Sample Surveys</h1>

<strong>Idea 1: Examining a Part of the Whole</strong>

-We would like to know about an entire group of individuals--a <strong>population</strong>--but examining all of them is usually impractical.  So we settle for examining a smaller group of individuals--a <strong>sample</strong>--selectted from the population.

-<strong>Bias</strong>: Selecting a sample to represent the population fairly is more difficult than it actually sounds.  Polls or surveys most often fail because they use a sampling method that tends to over or underrepresent parts of the population.  Sampling methods that, by their nature, tend to over or underemphasize some characteristics of the population are said to be biased.

-Modern surveys should always select individuals randomly.

<strong>Idea 2: Randomize</strong>

-Randomizing a sample protects us from the influences of all features of our population by making sure that, on average, the sample looks like the rest of the population.

<strong>Idea 3: Sample Size</strong>

-How large a random sample do we need for the sample to be reasonably representative of the population?  All that matters is the number of individuals in the sample.  A random sample of 100 students in a college represents the student body just about as well as a random sample of 100 voters represents the entire electorate of the United States.  The sample as a fraction of the population is not important, it's entirely about the number of individuals in the sample.

*<strong>Census</strong>: A census is a 'sample' of the entire population.  A census can be very difficult to complete, whether that be a result of population size or other factors.  Second, populations rarely stand still.  Third, taking a census can be more complex (and as a result more innacurate) than just sampling.

<h1 style="font-family:Courier;font-weight:bold;font-size:20px;">Simple Random Samples</h1>

-There are many ways to give everyone give everyone in a population an equal chance of being selected, but it is still not necassarily a representative sample.

*<strong>Simple Random Sample</strong>: Each combination of people has an equal chance of being selected.

-To select a sample at random, we first need to define where the sample will come from.  The <strong>sampling frame</strong> is a list of individuals from which the sample is drawn.  In defining the sample frame, we must deal with the details of defining the population.

-Once we have a sampling frame, the easiest way to choose an SRS is to assign a random number to each individual in the sampling frame.  We then select only those whose random numbers satisfy some rule.

*<strong>Sampling Variability</strong>:Samples drawn at random generally differ one from another.  Each draw of random numbers selects different people for our sample.  These differences lead to different values for the variables we measure.  We call these sample-to-sample differences <strong>sampling variability</strong>

<h1 style="font-family:Courier;font-weight:bold;font-size:20px;">Other Sampling Designs</h1>

-All statistical sampling designs have in common the idea that chance, rather than human choice, is used to select the sample.

<strong>Stratified Sampling</strong>

-Sometimes the population is first sliced into homogenous groups, called <strong>strata</strong> before the sample is selected.  Then simple random sampling is used within each stratum before the results are combined.  This common sampling design is called <strong>stratified random sampling</strong>.  This method helps us reduce sampling variability.  Stratified sampling can also help us notice important differences among groups.

<h1 style="font-family:Courier;font-weight:bold;font-size:20px;">Cluster and Multistage Sampling</h1>

-Splitting the population into representative <strong>clusters</strong> can make sampling more practical.  Then we could simply select one or a few clusters at random and perform a census within each of them.  This sampling design is called <strong>cluster sampling</strong>.  Clusters are generally selected for reasons of efficiency, practicality and cost.

-Sampling schemes that combine several methods are called <strong>multistage samples</strong>.

<h1 style="font-family:Courier;font-weight:bold;font-size:20px;">Systematic Samples</h1>

*<strong>Systematic Sampling</strong>:When using a systematic sampling method, you should justify the assumption that the systematic method is not associated with any of the measured variables.

<h1 style="font-family:Courier;font-weight:bold;font-size:20px;">From the Population to the Sample</h1>

-At each step in the sampling process, the group may be constrained further and further.  Each constraint can introduce more biases.  Whenever you work with or learn about a sample, pay attention not only to who was sampled bout to who was left out of the study.  The <em>who</em> in and SRS is the population from which we've drawn a representative sample.

<h1 style="font-family:Courier;font-weight:bold;font-size:20px;">The Valid Survey</h1>

-A valid survey yields the information we are seeking about the population we are interested in.

<h1 style="font-family:Courier;font-weight:bold;font-size:20px;">Common Sampling Mistakes</h1>

<strong>Mistake 1: Sample Volunteers</strong><br>
-A voluntary response sample can be a very bad idea.  The respondents, rather than the researcher decide who will be in the sample.  It's often hard to define the sampling frame of a voluntary response study.  Within the sampling frame, voluntary response samples are often biased toward those with strong opinions.  People with very negative opinions tend to respond more frequently than those with equally strong positive opinions.  The resulting <strong>voluntary response bias</strong> invalidates the survey.

<strong>Mistake 2: Sample Conveniently</strong><br>
-It is a bad idea to include individuals, simply because it is convenient for the sample.  This group may not be representative of the population.

<strong>Mistake 3: Use a Bad Sampling Frame</strong><br>
-An SRS from an incomplete sampling frame introduces bias because the individuals included may differ from the ones not in the frame.

<strong>Mistake 4: Undercoverage</strong><br>
-Many surveys suffer from this, some portion of the population is not sampled at all or has a smaller representation in the sample than it has in the population.

<h1 style="font-family:Courier;font-weight:bold;font-size:20px;">Nonresponse Bias</h1>

-It is a problem that those who don't respond to a survey, may drastically differ from those who do.  The lack of response will bias the results.  It is often best to design a smaller randomized survey for which the response rate will be higher.

<h1 style="font-family:Courier;font-weight:bold;font-size:30px;">Chapter12:Experiments and Observational Studies</h1>

<h1 style="font-family:Courier;font-weight:bold;font-size:20px;">Cluster and Multistage Sampling</h1>

*<strong>Observational Studies</strong>: In observational studies, researchers don't assign choices; they simply observe them.  

*<strong>Retrospective Study</strong>: Researchers identify subjects who are characterized by the independent variable, and then collect data about them on the response/dependent variable.

-Observational studies are valuable for discovering trends and possible relationships.  Observational studies that try to discover variables related to rare outcomes are often retrospective.  Retrospective studies, are however restrictive in their view in that they are usually limited to a small part of the entire population.  Also, because retrospective records are typically based on historical data, they can have errors.

-A more reliable approach, is to observe individuals over time recording the variables of interest and ultimately seeing how things turn out.  

*<strong>Prospective Study</strong>: If we identify subjects in advance and collect data as events unfold, we have a prospective study.

<h1 style="font-family:Courier;font-weight:bold;font-size:20px;">Randomized, Comparative Experiments</h1>

-An experiment requires a random assignment of subjects to treatments.  Experiments study the relationship between two or more variables.  An experimenter must identify at least one explanatory variable, called a <strong>factor</strong> to manipulate and at least one <strong>response variable</strong> to measure.

-What distinguishes the <strong>experiment</strong> from other types of investigation is that the experimenter actively and deliberately manipulates the factors to control the details of the possible treatments, and assigns the subjects to those treatments at random.

-Individuals on whom or which we experiment are known by a variety of terms.  When they are human, we usually call them <strong>subjects</strong> or <strong>participants</strong>.  Other non-human types are typically called <strong>experimental units</strong>

*<strong>Levels</strong>:The specific values that the experimenter chooses for a factor.

-The combination of specific levels from all the factors that an experimental unit receives is known as its <strong>treatment</strong>.

-In an experiment, it is imperative that treatments are assigned randomly.

<h1 style="font-family:Courier;font-weight:bold;font-size:20px;">The Four Principals of Experimental Design</h1>

-The significant need to control or minimize variation in an experiment!!

<ol>
    <li><strong>Control</strong><ul>
        <li>We control sources of variation other than the factors we are testing by making conditions as similar as possible for all treatment groups.</li>
        <li>Controlling extraneous sources of variation reduces the variability of the responses, making it easier to detect differences among the treatment groups.</li>
    </ul></li>
    <li><strong>Randomize</strong><ul>
        <li>Randomization allosw us to equalize the effects of unknown or uncontrollable sources of variation.  It does not eliminate the effects of these sources, but by distributing them equally (on average) across the treatment levels, it makes comparisons among the treatments fair.</li>
        <li>"Control what you can, and randomize the rest"</li>
    </ul></li>
    <li><strong>Replicate</strong><ul>
        <li>Drawing conclusions about the world is impossible unless we repeat, or replicate the results.</li>
        <li>Each treatment should be applied to a number of subjects, only with such replication can we estimate the variability of responses.</li>
        <li>A second kind of replication occurs when the entire experiment is repeated on a different population of experimental units.</li>
    </ul></li>
    <li><strong>Block</strong><ul>
        <li>Some attributes of the experimental units that we are not studying and that we can't control may nevertheless affect the outcomes of an experiment.</li>
        <li>If we group similar individuals together and then randomize within each of these blocks, we can account for much of the variability due to the difference among the blocks so it won't obscure our comparison of the treatment groups.</li>
        <li>Blocking is an important compromise between randomization and control</li>
    </ul></li>
</ol>

<h1 style="font-family:Courier;font-weight:bold;font-size:20px;">Does this make a difference?</h1>

-If the differences between the treatment groups are big enough, we'll attribute the differences to the treatments.  How do we decide whether or not the differences are big enough?

-The question should importantly be rephrased "Are the differences we observed between treatment groups about as big as we might get just from the randomization alone, or whether they're bigger than that."

-If the differences are bigger, we would say that the differences due to treatments are <strong>statistically significant</strong>.

-A good way to get a feeling for how significant the differences between treatments are is to examine the boxplots of each treatment group.  When the spreads are large, the observed difference may be just from random fluctuation.

<h1 style="font-family:Courier;font-weight:bold;font-size:20px;">Experiments and Samples</h1>

-Both experiments and sample surveys use randomization to get unbiased data, but they do so in different ways and for different purposes.

-Sample surveys try to estimate population parameters, so the sample needs to be as representative of the population as possible.

-By contrast, experiments try to assess the effects of treatments.  Experimental units are not always drawn randomly from the population.

-Unless the experimental units are chosen from the population at random, you should be cautious about generalizing experiment results to larger populations until the experiment has been repeated under different circumstances.

-Experiments can draw stronger conclusions than surveys.  By looking only at the differences across treatment groups, experiments cancel out many sorces of bias.

<h1 style="font-family:Courier;font-weight:bold;font-size:20px;">Control Treatments</h1>

*<strong>Control Treatment</strong>: The baseline measurement on experimental units, without any intentionally introduced treatment.  The experimental units within this group are called the <strong>control group</strong>.

<h1 style="font-family:Courier;font-weight:bold;font-size:18px;">Blinding</h1>

-Humans are notoriously susceptible to errors in judgement.  When we know what treatment was assigned, it's difficult not to let that knowledge influence our assessment of the response.

*<strong>Blinding</strong>: Masking the known variables in the equation.  When just the subjects in the study are blinded (class 1) the result is considered <strong>single-blind</strong>.  When the evaluators (class 2) are blinded as well, it is considered a <strong>double-blind</strong> experiment.

<h1 style="font-family:Courier;font-weight:bold;font-size:18px;">Placebos</h1>

-Often, simply applying any 'treatment' can induce an improvement in results.  Also, regression to the mean may account for some patients' improvement.  To seperate the effect of the treatment from these other effects we can use a control treatment that mimics the treatment itself.

*<strong>Placebo</strong>: A fake treatment that looks just like the treatment being tested.  Improvement in a treatment group controlled with the placebo is called the placebo effect, and is extremely common.

-In summary, the best experiments are usually:
<ul>
    <li>Randomized</li>
    <li>Comparative</li>
    <li>Double-blind</li>
    <li>Placebo-controlled</li>
</ul>

<h1 style="font-family:Courier;font-weight:bold;font-size:20px;">Blocking</h1>

-When groups of experimental units are similar, it's often a good idea to gather them together into blocks and then randomly assign the experimental units within each block to the treatments.  By blocking, we isolate the variability attributable to the differences between the blocks, so that we can see the differences caused by the treatments more clearly.

*<strong>Randomized Block Design</strong>: Random treatment assignment within each block group.

<h1 style="font-family:Courier;font-weight:bold;font-size:20px;">Confounding</h1>

*<strong>Confounded</strong>: When the levels of one factor are associated with the levels of another factor, we say that these two factors are confounded.

-Lurking variables and confounding variables are alike in that they interfere with our ability to interpret our analyses simply.  A lurking variable creates an association between two other variables that tempts us to think that one may cause the other.  A lurking variable is therefore thought of as a variable associated with both y and x that makes it appear that x may be causing y.

-A confounding variable, is associated in a noncausal way with a factor and affects the response.  Because of confounding, we find that we can't tell whether any effect we see was caused by our factor or by the confounding variable, or even by both working together.
